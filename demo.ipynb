```python
# Automaton of the algorithm as a Markov map
A.cf_automaton().plot()
```

```python
# Iterator generating the language of the automaton, which means the admissible products of matrices of the cocycle
# Input: n = length, i.e. number of states of a path in the automaton
# Option: length_n = 0 or 1 (default). If 1 it gives only the paths of length exactly n, if 0 all paths of length <= n
# Output: list of states the path is going through + the corresponding product matrix product

list(A.automaton_lang_mat_it(3))
```

```python
list(A.automaton_lang_mat_it(4))
```

```python
# cf_cylinders applies the admissible matrix products to the base subsimplices, obtaining a finer partition

A.cf_cylinders(3)
```

```python
# Helpful function localize_sim: it shows e.g. where a chosen simplex is in the cylinder decomposition
# Input: a product of matrices + a simplex. Output: It applies the matrix product to the simplex and it highlights it

A.cf_cylinders(3) + A.localize_sim(A._mat[0]*A._mat[0],[vector(x) for x in A._sim[0].rays()])
```
